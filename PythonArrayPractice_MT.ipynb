# * 

```python
from array import *
my_array = array("i",[1,3,5,7,9])

print("Current memory address and length, in elements, of the buffer: ", str(my_array.buffer_info()))
print("Current memory size: ", str(my_array.buffer_info()[1] * my_array.itemsize))
```
