```python
#create graph
my_graph = {"Sonic" : ["Tails","Knuckles"],
           "Tails": ["Knuckles","Cream"],
           "Knuckles":["Rouge","Blaze"],
           "Rouge": ["Knuckles"]}
```

```python
#Algorithm for finding path from one vertice to another in graph

def find_path(my_graph,start,end, path = []):
    path = path + [start]
    if start == end:
        return path
    if start not in my_graph:
        return None
    for node in my_graph[start]:
        if node not in path:
            newpath = find_path(start, node, end,path)
            if newpath: return newpath
    print("No path avaliable")
```

```python
#test out with sample path
find_path(my_graph, "Sonic","Rouge")
find_path(my_graph, "Sonic", "Knuckles")
```
